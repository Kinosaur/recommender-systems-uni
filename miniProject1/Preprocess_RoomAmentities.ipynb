```python
import pandas as pd
import re

df = pd.read_csv('HOTEL_OUTDATASET.csv')
mapping = {
    'free wifi in all rooms': 'wifi',
    'inroom tablet': 'tablet',
    'mini bar': 'minibar',
    'inroom safe box': 'safebox',
    'tv flat screen': 'tv',
    'tv in bathroom': 'tv',
    'wifi in public areas': 'public wifi',
    'pets allowed in room':'petallow',
    'interconnecting rooms available':'interconnecting rooms '
}

def normalize_with_dict(amenity):
    key = amenity.lower().strip()
    key = re.sub(r'[^\w\s]', '', key)  
    return mapping.get(key, key)

all_amenities = df['roomamenities'].dropna().str.split(';').sum()

normalized_amenities = sorted(set(normalize_with_dict(a.strip()) for a in all_amenities))

print("Unique normalized amenities:")
for amenity in normalized_amenities:
    print(amenity)
```

```python
import pandas as pd

df = pd.read_csv('HOTEL_OUTDATASET.csv')
raw_mapping = {
    'free wifi in all rooms': 'wifi',
    'in-room tablet': 'tablet',
    'mini bar': 'minibar',
    'in-room safe box': 'safebox',
    'tv flat screen': 'tv',
    'tv in bathroom': 'tv',
    'wifi in public areas': 'public wifi',
    'pets allowed in room':'petallow',
    'interconnecting rooms available':'interconnecting rooms',
    'satellitecable channels':'satellitecable'
}
mapping = {
    k.lower().strip(): v.lower().strip()
    for k, v in raw_mapping.items()
}
def normalize_amenity(amenity) -> str:
    key = amenity.lower().strip()
    return mapping.get(key, key)

def amenities_to_vector(amenities_str: str) -> set:
    if pd.isna(amenities_str):
        return set()
    raw_items = [a.strip() for a in amenities_str.split(';') if a.strip()]
    normalized = {normalize_amenity(a) for a in raw_items}
    return normalized

df['amenity_vector'] = df['roomamenities'].apply(amenities_to_vector)

for i in range(min(5, len(df))):
    print(f"Room {i+1}: {df['amenity_vector'].iloc[i]}")
```
