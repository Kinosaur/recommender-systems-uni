```python
import pandas as pd
import numpy as np
```

```python
datas = pd.read_csv('full_book_details.csv', encoding='latin-1')
datas.head()
```

```python
datas['name'].head()
```

```python
user_rating = pd.pivot_table(data=datas, values='user_rating', index='user_id', columns='name')
user_rating.head()
```

```python
rating = pd.DataFrame(datas.groupby('name')['no_of_ratings'].mean())
rating
```

```python
rating.sort_values('no_of_ratings', ascending=False).head(10)
```

# Choose 1 book and get reccomendation based on that

```python
title = "The Hiding Place: The Triumphant True Story of Corrie Ten Boom"
book_user_ratings = user_rating[title]
book_user_ratings.head()
```

```python
similar_to_book = user_rating.corrwith(book_user_ratings)
similar_to_book.head()
```

```python
corr_book = pd.DataFrame(similar_to_book, columns=['Correlation'])
corr_book.dropna(inplace=True)
```

```python
corr_book.head()
```

```python
corr_book.sort_values('Correlation', ascending=False).head(10)
corr_book = corr_book.join(rating['no_of_ratings'])
corr_book.head()
```
